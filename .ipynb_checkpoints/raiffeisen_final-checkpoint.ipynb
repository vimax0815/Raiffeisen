##  3. Генерация признаков<a id="3"></a>
[Отбор признаков▼](#4) [Оглавление▲▲](#heading)

```python
group_city = df.groupby('city')['total_square'].agg({'mean'}).reset_index()

def group_feat_city(row):
    c = row['city'] 
    row['mean_group_city'] = group_city.loc[group_city['city']==c, 'mean'].iloc[0]
#     row['min_group_city'] = group_city.loc[group_city['city']==c, 'min'].iloc[0]
#     row['max_group_city'] = group_city.loc[group_city['city']==c, 'max'].iloc[0]
    return row

df = df.apply(lambda x: group_feat_city(x), axis=1)

df['total/mean_square_city'] = df['total_square'] / df['mean_group_city']
```

```python
group_reg = df.groupby('region')['total_square'].agg({'mean'}).reset_index()

def group_feat_reg(row):
    r = row['region'] 
    row['mean_group_reg'] = group_reg.loc[group_reg['region']==r, 'mean'].iloc[0]
#     row['min_group_reg'] = group_reg.loc[group_reg['region']==r, 'min'].iloc[0]
#     row['max_group_reg'] = group_reg.loc[group_reg['region']==r, 'max'].iloc[0]
    return row

df = df.apply(lambda x: group_feat_reg(x), axis=1)

df['total/mean_square_reg'] = df['total_square'] / df['mean_group_reg']
```

```python
group_type = df.groupby('realty_type')['total_square'].agg({'mean'}).reset_index()

def group_feat_type(row):
    t = row['realty_type'] 
    row['mean_group_type'] = group_type.loc[group_type['realty_type']==t, 'mean'].iloc[0]
#     row['min_group_type'] = group_type.loc[group_type['realty_type']==t, 'min'].iloc[0]
#     row['max_group_type'] = group_type.loc[group_type['realty_type']==t, 'max'].iloc[0]
    return row

df = df.apply(lambda x: group_feat_type(x), axis=1)

df['total/mean_square_type'] = df['total_square'] / df['mean_group_type']
```

##  4. Отбор признаков<a id="4"></a>
[Прогнозирование таргета▼](#5) [Оглавление▲▲](#heading)

```python
y = np.log(df.loc[mask, 'target'])
df = df.drop('target', axis=1)
cat_col = df.select_dtypes(include=['object']).columns.to_list()
num_col = df.select_dtypes(exclude=['object']).columns.to_list()
```

```python
X = df[mask]
X.shape, y.shape
y_shap = np.exp(y)
```

```python
model = CatBoostRegressor(random_state=42)
model.fit(X, y_shap, cat_col,verbose=False, plot=False)

explainer = shap.Explainer(model)
shap_values = explainer(X)
```

```python
shap.plots.bar(shap_values, max_display=100)
```

```python
imp = pd.DataFrame(abs(shap_values.values), columns=X.columns).mean(axis=0).sort_values(ascending=False).reset_index()
imp
```

```python
imp = imp[:80]
imp_list = list(imp['index'])
#imp_list
```

```python
df = df[imp_list]
df.head(2)
```

##  5. Прогнозирование таргета<a id="5"></a>
[Оглавление▲▲](#heading)

```python
cat_col = df.select_dtypes(include=['object']).columns.to_list()
num_col = df.select_dtypes(exclude=['object']).columns.to_list()
```

```python
train = df[mask]
test = df[~mask]
```

```python
scaler = MinMaxScaler()
train[num_col] = scaler.fit_transform(train[num_col])
test[num_col] = scaler.transform(test[num_col])
train.shape, y.shape
```

```python
train.head(2)
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(train, y, test_size=0.25, random_state=42)
x_train.shape, y_train.shape
x_valid.shape, y_valid.shape
```

```python
cat = CatBoostRegressor(random_state=42)
cat.fit(X=x_train, y=y_train, eval_set=(x_valid, y_valid), cat_features=cat_col, verbose=0)
cat.get_best_score()
```

```python
cat.fit(X=train, y=y, cat_features=cat_col, verbose=0)
pred = np.exp(cat.predict(test))
```
